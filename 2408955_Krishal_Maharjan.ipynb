#Task - 3 - Outlier Detection:
Tasks:

1. Identify outlier countries in South Asia based on their Score and GDP per Capita.

2. Define outliers using the 1.5 × IQR rule.

3. Create a scatter plot with GDP per Capita on the x-axis and Score on the y-axis, highlighting outliers
in a different color.

4. Discuss the characteristics of these outliers and their potential impact on regional averages.

```python
# Step 1: Calculate IQR for Score and GDP per Capita
def detect_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] < lower_bound) | (df[column] > upper_bound)]


# Detect outliers for 'score' and 'GDP per capita'
score_outliers = detect_outliers(df2,'score')
gdp_outliers = detect_outliers(df2,'Log GDP per capita')

# Combine both outlier sets to identify all unique outliers
outliers = pd.concat([score_outliers, gdp_outliers]).drop_duplicates()

# Scatter Plot with Outliers Highlighted
plt.figure(figsize=(10, 3))

sns.scatterplot(
    data=south_asia,
    x='Log GDP per capita',
    y='score',
    color='green',
    label='Normal'
)
sns.scatterplot(
    data=outliers,
    x='Log GDP per capita',
    y='score',
    color='blue',
    label='Outliers'
)

plt.title('GDP per Capita vs. Score with Outliers Highlighted', fontsize=16)
plt.xlabel('GDP per Capita', fontsize=12)
plt.ylabel('Happiness Score', fontsize=12)
plt.legend()
plt.show()
```

3. Metric Comparisons:
• Compare key metrics like GDP per Capita, Social Support, and Healthy Life Expectancy
between the regions using grouped bar charts.
• Which metrics show the largest disparity between the two regions?

```python
# Key metrics for comparison
metrics = ['Log GDP per capita', 'Social support', 'Healthy life expectancy']

# Create deep copies to avoid SettingWithCopyWarning
south_asia_metrics = df2[['Country name'] + metrics].copy()
middle_east_metrics = middle_east[['Country name'] + metrics].copy()

# Add 'Region' column to each dataframe
south_asia_metrics['Region'] = 'South Asia'
middle_east_metrics['Region'] = 'Middle East'

# Combine the dataframes and reset the index to avoid duplicate labels
combined_metrics = pd.concat([south_asia_metrics, middle_east_metrics], ignore_index=True)

# Melting the data for easier plotting of grouped bar charts
melted_metrics = combined_metrics.melt(
    id_vars=['Country name', 'Region'],
    value_vars=metrics,
    var_name='Metric',
    value_name='Value'
)

# Plotting grouped bar chart
plt.figure(figsize=(14, 8))
sns.barplot(
    x='Metric',
    y='Value',
    hue='Region',
    data=melted_metrics,
    palette='Blues'
)
plt.title('Key Metrics Comparison Between South Asia and Middle East')
plt.xlabel('Metric')
plt.ylabel('Value')
plt.xticks(rotation=45)
plt.legend(title='Region')
plt.show()
```

4. Happiness Disparity:
• Compute the range (max - min) and coefficient of variation (CV) for Score in both regions.
• Which region has greater variability in happiness?


```python
# Range and Coefficient of Variation (CV) for Score in both regions
south_asia_range = df2['score'].max() - df2['score'].min()
middle_east_range = middle_east['score'].max() - middle_east['score'].min()

south_asia_cv = df2['score'].std() / df2['score'].mean()
middle_east_cv = middle_east['score'].std() / middle_east['score'].mean()

print(f"South Asia - Range: {south_asia_range}, \nCV: {south_asia_cv}\n")
print(f"Middle East - Range: {middle_east_range}, \nCV: {middle_east_cv}")

if south_asia_cv > middle_east_cv:
    print("\nSouth Asia has greater variability in happiness.")
else:
    print("\nMiddle East has greater variability in happiness.")
```
